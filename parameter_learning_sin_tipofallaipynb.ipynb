```
import pandas as pd
url= 'https://raw.githubusercontent.com/karlajagg/Documentacion/main/xtrainfallasf3.csv'
data=pd.read_csv(url)
print (data)
```

```
data1 = data[['ST','EST','CIM', 'BRE', 'NOMS', 'CHC', 'COM', 'AB', 'ACC', 'CHH','LC']]
print(data1)
```

```
from pgmpy.models import BayesianModel

model = BayesianModel([('ST', 'CIM'),('ST', 'COM'),('CIM', 'EST'),('EST', 'AB'),('AB', 'COM') , ('CIM','NOMS'),  ('NOMS','COM'),  ('NOMS','CHC'),  ('NOMS','ACC'),  ('CHC','COM'),  ('CHC','AB'),  ('CHC','CHH'),  ('LC','NOMS'),  ('BRE','EST'),  ('BRE','CHH'),  ('CHH','ACC'),  ('ACC','AB')]) 
#El aprendizaje de parámetros es la tarea de estimar los valores de las 
#distribuciones de probabilidad condicional (CPD),
```

```
from pgmpy.estimators import ParameterEstimator
pe = ParameterEstimator(model, data)
#CONTADOR DE ESTADOS
print("\n", pe.state_counts('ST')) 
print("\n", pe.state_counts('EST')) 
print("\n", pe.state_counts('CIM')) 
print("\n", pe.state_counts('NOMS')) 
print("\n", pe.state_counts('LC')) 
print("\n", pe.state_counts('BRE'))
print("\n", pe.state_counts('COM'))  
print("\n", pe.state_counts('CHC'))
print("\n", pe.state_counts('CHH')) 
print("\n", pe.state_counts('AB'))
print("\n", pe.state_counts('ACC'))
```

```
#Maximum Likelihood Estimation
from pgmpy.estimators import MaximumLikelihoodEstimator
mle = MaximumLikelihoodEstimator(model, data)
print(mle.estimate_cpd('ST')) # conditional
print(mle.estimate_cpd('EST')) # conditional
print(mle.estimate_cpd('CIM'))  # unconditional
print(mle.estimate_cpd('NOMS')) # conditional
print(mle.estimate_cpd('LC')) # unconditional
print(mle.estimate_cpd('BRE')) # unconditional
print(mle.estimate_cpd('COM'))  # conditional
print(mle.estimate_cpd('CHC')) # conditional
print(mle.estimate_cpd('CHH')) # unconditional
print(mle.estimate_cpd('AB')) # unconditional
print(mle.estimate_cpd('ACC')) # unconditional
```

```
# Calibrate all CPDs of `model` using MLE:
model.fit(data, estimator=MaximumLikelihoodEstimator)
```

```
from pgmpy.estimators import BayesianEstimator
# BayesianEstimator, too, can be used via the fit()-method
# Aqui imprime todas las CPDs
model.fit(data, estimator=BayesianEstimator, prior_type="BDeu", equivalent_sample_size=10) # default equivalent_sample_size=5
for cpd in model.get_cpds():
    print(cpd)
```
